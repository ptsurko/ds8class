```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'key1': ['a','a','b','b','a'],
                   'key2': ['one', 'two', 'one', 'two', 'one'],
                   'data1': np.random.randn(5),
                   'data2': np.random.randn(5)})
df
```

```python
df.groupby(['key1', 'key2'])[['data1']].mean()
```

```python
people = pd.DataFrame(np.random.randn(5,5),
                      columns=['a','b','c','d','e'],
                      index=['Joe','Steve','Wes','Jim','Travis'])
mapping = {'a':'red','b':'red','c':'blue','d':'blue','e':'red','f':'orange'}

people.groupby(mapping, axis=1).sum()
```

```python
people.groupby(len).sum()
```

```python
key_list=['one','one','one','two','two']
people.groupby([len, key_list]).min()
```

```python
tips = pd.read_csv('examples/tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
grouped = tips.groupby(['day', 'smoker'])
grouped_pct = grouped['tip_pct']
grouped_pct.agg([('a', 'mean'), 'std'])
```

```python
functions = ['count','mean','max']
grouped['tip_pct', 'total_bill'].agg(functions)
```

```python
grouped.agg({'tip': [np.max, np.min, 'mean', 'std'], 'size': 'sum'})
```

```python
tips.groupby(['day','smoker'], as_index=False).mean()
```

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by='tip_pct', ascending=False)[:n]

tips.groupby(['smoker', 'day']).apply(top, n=2)
```
