```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams['figure.figsize'] = [15, 8]

data = np.arange(10)
data

# plt.plot(data)

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
plt.plot(np.random.randn(50).cumsum(), 'k-x')
```

```python
fix, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=500, color='k', alpha=0.5)
        
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'])
ax.set_title('My first matplot')
ax.set_xlabel('Stages')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum(), 'r', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'g--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'b.', label='three')
ax.legend(loc='best')
```

```python
tips = pd.read_csv('tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
# party_counts = party_counts.loc[:,2:5]

party_counts = party_counts.div(party_counts.sum(1), axis=0)
party_counts.plot.bar()
```

```python
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
sns.set(style='whitegrid')
sns.barplot(x='tip_pct', y='day', data=tips, hue='time', orient='h')
```
